## If-else statement:ðŸ§¨

```python
# If-else statement
x = 5
if x > 10:
    print("x is greater than 10")
else:
    print("x is less than or equal to 10")

# If-elif-else statement
y = 15
if y < 10:
    print("y is less than 10")
elif 10 <= y < 20:
    print("y is between 10 and 19")
else:
    print("y is greater than or equal to 20")
```

## For loop exampleâ›“

```python
# For loop example
print("For loop:") # Display numbers one to five Using the for command
for i in range(1, 6):
    print(i)

# While loop example
print("\nWhile loop:") # Display numbers one to five Using the While command
j = 1
while j < 6:
    print(j)
    j += 1

# Loop that iterates over a list : Looping over the elements of a list
fruits = ["apple", "banana", "orange"]
print("\nLoop that iterates over a list:")
for fruit in fruits:
    print(fruit)
print("List of fruits:", fruits)
```

## Create a list of integersâœ¨

```python
# Create a list of integers
my_list = [1, 2, 3, 4, 5]

# Accessing elements of the list
print("First element:", my_list[0])  # Output: 1
print("Third element:", my_list[2])  # Output: 3
print("Last element:", my_list[-1])  # Output: 5

# Slicing the list
print("Sliced sublist:", my_list[1:3])  # Output: [2, 3]
print("Sliced sublist:", my_list[:4])  # Output: [1, 2, 3, 4]

# Adding elements to the list
my_list.append(6) # add
my_list.insert(0, "Asghar") # to replace

# Changing list elements
my_list[1:4] = ["Mammad", "Alice", "Ali"]
print("Updated list:", my_list)

# Removing an element from the list
my_list.remove("Mammad")
print("List after removal:", my_list)

# Length of the list
print("Length of the list:", len(my_list))
```

## Define a function ðŸ“š

```python
# Define a function to greet a person by name
def greet(name): # with a specific name :

    print("Hello, " + name)

# Call the greet function with a specific name
greet("Alice")

# Define a function with a default argument
def greet(name="world"): # with a default argument :

    print("Hello, " + name)

# Call the greet function with different names
greet()  # Prints "Hello, world"
greet("Alice")  # Prints "Hello, Alice"
```

```python
# Define a class for representing a person
class Person:
    def __init__(self, name, age):

        self.name = name
        self.age = age

    def say_hello(self):

        print("Hello, my name is " + self.name)

# Define a subclass called Student that inherits from Person
class Student(Person):
    def __init__(self, name, age, student_id):

        super().__init__(name, age)
        self.student_id = student_id

    def study(self, subject):

        print(self.name + " is studying " + subject)

# Create an instance of the Person class
person1 = Person("Alice", 25)

# Call a method on the Person instance to greet
person1.say_hello()

# Create an instance of the Student class
student1 = Student("Bob", 20, "12345")

# Call a method on the Student instance to greet
student1.say_hello()

# Call a method specific to the Student class
student1.study("Math")
```

## NumpyðŸ“™

```python
import numpy as np

# Creating NumPy arrays
arr1d = np.array([1, 2, 3, 4, 5])
arr2d = np.array([[1, 2, 3], [4, 5, 6]])
arr3d = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
print("arr2d:")
print(arr2d)
print("arr3d:")
print(arr3d)

# Array indexing
print("Array indexing:")
print("Element at index 3 in arr1d:", arr1d[3])
print("Element at row 1, column 2 in arr2d:", arr2d[1, 2])
print("Element at layer 1, row 0, column 1 in arr3d:", arr3d[0, 1, 0])

# Array slicing
print("\nArray slicing:")
print("Slice of arr1d from index 1 to 4:", arr1d[1:4])
print("Slices of arr2d from all rows, starting from column 1:", arr2d[:, 1:])

# Array operations
print("\nArray operations:")
arr = np.array([1, 2, 3])
print("Adding 2 to each element:", arr + 2)
print("Multiplying each element by 2:", arr * 2)
```

```python
import numpy as np

# Arithmetic operations
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

print("Addition of arrays:")
print(arr1 + arr2)

print("\nSubtraction of arrays:")
print(arr1 - arr2)

print("\nMultiplication of arrays:")
print(arr1 * arr2)

print("\nDivision of arrays:")
print(arr2 / arr1)

print("\nSquare of elements in arr1:")
print(arr1 ** 2)

# Statistical operations
arr = np.array([1, 2, 3, 4, 5])

print("\nMean of elements in arr:")
print(np.mean(arr))

print("\nMedian of elements in arr:")
print(np.median(arr))

print("\nVariance of elements in arr:")
print(np.var(arr))

print("\nMinimum value in arr:")
print(np.min(arr))

print("\nMaximum value in arr:")
print(np.max(arr))
```

```python
import numpy as np

# Concatenate two NumPy arrays along axis 0 (row-wise)
result = np.concatenate((arr1, arr2), axis=0)

# Print the concatenated result
print("Concatenated result along axis 0 (row-wise):")
print(result)
```

## Two-dimensional diagramðŸ“ˆ

```python
import matplotlib.pyplot as plt
import numpy as np

# Create some data
x = np.linspace(-10, 10, 500)
y = np.sinc(x)

# Plot the data
plt.plot(x, y, label="Sine Wave", color="red", linewidth=2)  # Adding labels and style
plt.xlabel("X-axis")  # Label for the x-axis
plt.ylabel("Y-axis")  # Label for the y-axis
plt.title("Sinc Wave Plot")  # Title for the plot
plt.grid(True)  # Display grid lines (optional)
plt.legend(loc="upper right")  # Add a legend (optional)
plt.show()
```

## Display multiple charts on one page

```python
import matplotlib.pyplot as plt
import numpy as np

# Create some data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create a figure and axis object
fig, ax = plt.subplots()  # Set the figure size (optional)

# Plot the data with labels and styles
ax.plot(x, y1, label="Sin", color="blue", linewidth=2)
ax.plot(x, y2, label="Cos", color="red", linestyle="--", linewidth=2)

# Add labels to axes
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")

# Add a title to the plot
ax.set_title("Sine and Cosine Waves")

# Display a legend
ax.legend(loc="upper right")

# Display grid lines (optional)
ax.grid(True)

# Show the plot
plt.show()
```

## Bar graphðŸ“Š

```python
import matplotlib.pyplot as plt
import numpy as np

# Create some data
labels = ["A", "B", "C", "D", "E"]
values = [20, 35, 30, 10, 5]

# Create a bar plot
plt.bar(labels, values, color='blue')  # Customize bar color
plt.xlabel('Categories')  # Label for the x-axis
plt.ylabel('Values')  # Label for the y-axis
plt.title('Bar Plot of Categories and Values')  # Title for the plot

# Display values on top of bars
for i, v in enumerate(values):
    plt.text(i, v + 1, str(v), ha='center', va='bottom')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data points following a normal distribution
x = np.random.normal(size=1000)

# Create a histogram
plt.hist(x, bins=30, color='red', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Random Data')
# Show the plot
plt.show()
```

```python
import pandas as pd

# Create a DataFrame
data = {
    "Name": ["sara", "Ali", "reza", "Iman"],
    "Age": [25, 30, 35, 40],
    "Gender": ["F", "M", "M", "M"]
}

df = pd.DataFrame(data)

# Add a new column to the DataFrame
df["Salary"] = [50000, 60000, 70000, 80000]

# Group the data by gender and calculate the mean age and salary
grouped = df.groupby("Gender")
result = grouped.agg({"Age": "mean", "Salary": "mean"})

# Print the result
print("Mean Age and Salary Grouped by Gender:")
print(result)
```
